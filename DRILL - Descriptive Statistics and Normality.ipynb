```python
# DRILL - Descriptive Statistics and Normality

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statistics
```

```python
binomial = np.random.binomial(50, 0.5, 100)
plt.hist(binomial)
plt.title('Binomial Distribution')
plt.xlabel('Random Values')

def plotter(binomial):
    plt.axvline(binomial.mean(), color = 'r', linestyle='solid', linewidth=2)
    plt.axvline(binomial.mean()+binomial.std(), color='r', linestyle='dashed', linewidth=2)
    plt.axvline(binomial.mean()-binomial.std(), color='r', linestyle='dashed', linewidth=2)
plotter(binomial)
plt.show()

def stats(binomial):
    print('mean: ', np.mean(binomial))
    print('standard deviation: ', np.std(binomial))
stats(binomial)
```

### Binomial Distribution – this is a bell shaped distribution with roughly half the instances on each side of the mean.  
### The descriptive statistics are useful for this distribution, which is used for single occurence tests.

```python
pareto = np.random.pareto(3, 1000)
plt.hist(pareto)
plt.title('Pareto Distribution')
plt.xlabel('Random Values')

plotter(pareto)
plt.show()

stats(pareto)
```

### Pareto Distribution – the 80-20 rule, an example is distribution of wealth, with values congregated near the mean.  
### The descriptive statistics are less useful since the distribution is skewed so dramatically.

```python
chisquare = np.random.chisquare(3, 1000)
plt.hist(chisquare)
plt.title('Chisquare Distribution')
plt.xlabel('Random Values')

plotter(chisquare)
plt.show()

stats(chisquare)
```

### Chisquare - chi-squared distribution is the square of a standard normal distribution.
### primarily used to determine whether there is a significant association between two variables.
### It measures how well the observed distribution of data fits with the distribution that is expected 
###    if the variables are independent.

```python
exponential = np.random.exponential(1, 1000)
plt.hist(exponential)
plt.title('Exponential Distribution')
plt.xlabel('Random Values')

plotter(exponential)
plt.show()

stats(exponential)
```

### Exponential Distribution – usually represents time until a given event occurs, thus is skewed left, 
###   then passes the mean as it tapers off.

```python
logistic = np.random.logistic(50, 1, 1000)
plt.hist(logistic)
plt.title('Logistic Distribution')
plt.xlabel('Random Values')

plotter(logistic)
plt.show()

stats(logistic)
```

### Logistic Distribution – this is similar to the normal distribution, with slightly longer tails.  
### The descriptive statistics fit well.

```python
laplace = np.random.laplace(80, 2, 1000)
plt.hist(laplace)
plt.title('Laplace Distribution')
plt.xlabel('Random Values')

plotter(laplace)
plt.show()

stats(laplace)
```

### Laplace Distribution – Back to back exponential distributions at mean, so naturally going to be congregated about mean.  
### The descriptive statistics are less useful.

```python
normal1 = np.random.normal(5, 0.5, 1000)
normal2 = np.random.normal(10, 1, 1000)
normal_joined = normal1 + normal2

plt.hist(normal_joined)
plt.title('Normal Distribution of combined variables')
plt.xlabel('Random Values')

plotter(normal_joined)
plt.show()

stats(normal_joined)
```

### Normal Distribution – 2 normal distributions joined together produce a normal distribution, 
### with similar descriptive statistics to individual distributions
