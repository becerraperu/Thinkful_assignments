```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('basket.csv')
df.head()
df.info()

# This data set contains statistics for 22,083 basketball players.
# These stats include games played, points, assists, rebounds etc.
```

```python
plt.scatter(x=df['GP'], y=df['points'])
plt.xlabel('Games played')
plt.ylabel('points')
plt.title('Games played / points')
plt.show()
```

### This scatter plot show games played vs points.
### There is general positive correlation - more games result in more points.
### Some players who played many games didn't necessarily score many points.


```python
df.corr()
```

### There is a strong correlation between points and attempts(.98).
### There is a medium correlation between points and blocks(.40).
### Some of the offensive players should get back on defense and block more.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(df['points'], color='purple')
plt.ylabel('points')
plt.title('Points made')

plt.subplot(1, 2, 2)
plt.plot(df['assists'], color='green')
plt.ylabel('assists')
plt.title('Assists')
plt.show()
```

### There were a few players that scored many points in their career, a sharp drop off, then the mid-tier players.
### The drop off in assists was less dramatic than that of points as the quality decreased.

```python
plt.hist(df['points'],bins=50)
plt.xlabel('number of points')
plt.ylabel('number of players')
plt.title('Points Histogram')
plt.show()

average = np.mean(df['points'])
std_dev = np.std(df['points'])

print('average number of points = ', round(average,2))
print('standard deviation = ', round(std_dev,2))
```

### The histogram of points shows that there were many players that made few points, and few players that made many points.
### This data show that players averaged about 487 points per year.


```python
plt.boxplot(df['points'])

median = np.median(df['points'])
print('The median number of points made per year =', median)
```

### The box plot shows that a bulk of the players made less than 1000 points.
### The number of outliers shows that superstars made a considerable amount of points more than the average player.
